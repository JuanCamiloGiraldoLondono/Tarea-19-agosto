```python
import string
import random
from typing import List, Callable


def generar_id(length=8):
    # helper function for generating an id
    return ''.join(random.choices(string.ascii_uppercase, k=length))


class Ticket_soporte:

    def __init__(self, customer, issue):
        self.id = generar_id()
        self.customer = customer
        self.issue = issue


def fifoOrdering(list: List[Ticket_soporte]) -> List[Ticket_soporte]:
    return list.copy()


def filoOrdering(list: List[Ticket_soporte]) -> List[Ticket_soporte]:
    list_copy = list.copy()
    list_copy.reverse()
    return list_copy


def randomOrdering(list: List[Ticket_soporte]) -> List[Ticket_soporte]:
    list_copy = list.copy()
    random.shuffle(list_copy)
    return list_copy


def blackHoleOrdering(list: List[Ticket_soporte]) -> List[Ticket_soporte]:
    return []


class Soporte_cliente:

    def __init__(self):
        self.tickets = []

    def Crear_ticket(self, customer, issue):
        self.tickets.append(Ticket_soporte(customer, issue))

    def Procesar_ticket(self, ordering: Callable[[List[Ticket_soporte]], List[Ticket_soporte]]):
        # create the ordered list
        Lista_ticket = ordering(self.tickets)

        # if it's empty, don't do anything
        if len(Lista_ticket) == 0:
            print("There are no tickets to process. Well done!")
            return

        # go through the tickets in the list
        for ticket in Lista_ticket:
            self.process_ticket(ticket)

    def process_ticket(self, ticket: Ticket_soporte):
        print("==================================")
        print(f"Procesando id tiwuete: {ticket.id}")
        print(f"cliente: {ticket.customer}")
        print(f"Issue: {ticket.issue}")
        print("==================================")


# create the application
app = Soporte_cliente()

# register a few tickets
app.Crear_ticket("John Smith", "Mi computador hace ruidos extra√±os!")
app.Crear_ticket("Linus Sebastian", "No puedo subir ningun video, ayuda.")
app.Crear_ticket("Arjan Egges", "VSCode no resuelve automaticamente mis problemas.")

# process the tickets
app.Procesar_ticket(blackHoleOrdering)
```
